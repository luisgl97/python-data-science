### Generar Datos a graficar

#### Examen de Certificacion para estudiantes de 5 niveles

```python
import numpy as np

años=["2011","2012","2013","2014","2015","2016","2017","2018","2019","2020"]

nivel1 = np.random.rand(10)*100
nivel2 = np.random.rand(10)*200 + 100
nivel3 = np.random.rand(10)*300 + 200
nivel4 = np.random.rand(10)*400 + 300
nivel5 = np.random.rand(10)*500 + 400
```

### Personalización de Gráficas

```python
# Introducir series de datos con color, marcador, etiqueta y estilo

plt.plot(años, nivel1, label="Nivel 1", color="purple", marker="<", linestyle="-")
plt.plot(años, nivel2, label="Nivel 2", color="red", marker="*", linestyle="--")
plt.plot(años, nivel3, label="Nivel 3", color="blue", marker="^", linestyle=":")
plt.plot(años, nivel4, label="Nivel 4", color="black", marker=".", linestyle="-.")
plt.plot(años, nivel5, label="Nivel 5", color="green", marker="+", linestyle="-")

# Activar leyendas
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

# Titulo de Grafica
plt.title("Examen de Certificación")

# Etiquetas de los ejes
plt.xlabel("Añis que se ha realizado el examen")
plt.ylabel("Puntaje")

# Personalizar Eje de puntaje de 0 a 1000 en incrementos de 100
plt.yticks(np.arange(0, 1001, 100))

# Activar cuadricula
plt.grid()

# Activar marcas menores
#plt.minorticks_on()

plt.show()
```

### Gráfica de Barras Verticales (bar)

```python
eje_x = np.arange(0,10)

plt.bar(eje_x, nivel5, label="Nivel 5", width=1/5)
plt.bar(eje_x + 0.2, nivel4, label="Nivel 4", width=1/5)
plt.bar(eje_x + 0.4, nivel3, label="Nivel 3", width=1/5)
plt.bar(eje_x + 0.6, nivel2, label="Nivel 2", width=1/5)
plt.bar(eje_x + 0.8, nivel1, label="Nivel 1", width=1/5)

plt.xticks(np.arange(0, 10, 1))
plt.yticks(np.arange(0,901,100))
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.show()
```

### Gráfica de Barras Horizontales (barh)

```python
plt.barh(eje_x + 0.8, nivel5, label="Nivel 5", height=1/5)
plt.barh(eje_x + 0.6, nivel4, label="Nivel 4", height=1/5)
plt.barh(eje_x + 0.4, nivel3, label="Nivel 3",height=1/5)
plt.barh(eje_x + 0.2, nivel2, label="Nivel 2",height=1/5)
plt.barh(eje_x, nivel1, label="Nivel 1",height=1/5)

plt.yticks(range(11))
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

### Grafica de Barras Verticales Apiladas

```python
plt.bar(eje_x, nivel5,label="Nivel 5",color="black")
plt.bar(eje_x, nivel4,label="Nivel 4", bottom=nivel5)
plt.bar(eje_x, nivel3,label="Nivel 3", bottom=nivel4+nivel5)
plt.bar(eje_x, nivel2,label="Nivel 2", bottom=nivel3+nivel4+nivel5)
plt.bar(eje_x, nivel1,label="Nivel 1", bottom=nivel2+nivel3+nivel4+nivel5)

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')

plt.show()
```

### Gráfica de Dispersión (scatter)

```python
plt.scatter(eje_x, nivel1, label="Nivel 1", marker="*")
plt.scatter(eje_x, nivel2, label="Nivel 2")
plt.scatter(eje_x, nivel3, label="Nivel 3")
plt.scatter(eje_x, nivel4, label="Nivel 4")
plt.scatter(eje_x, nivel5, label="Nivel 5")

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

### Gráfica de Pie

```python
plt.pie(nivel1,
        labels=["México", "Colombia","España","Estados Unidos", "Ecuador",
                "Argentina","Venezuela","Puerto Rico","Honduras","Uruguay"])

plt.show()
```

### Combinar Gráficas

```python
plt.bar(eje_x, nivel1, width=1/5, label="Nivel 1")
plt.bar(eje_x+0.2, nivel2, width=1/5, label="Nivel 2")

plt.plot(años, nivel3, label="Nivel 3")

plt.scatter(eje_x, nivel4, marker="*", label="Nivel 4")
plt.scatter(eje_x, nivel5, label="Nivel 5")

plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```
