## Generar arreglos con datos aleatorios

```python
import numpy as np

np.random.rand(10)
```

```python
#Distribucion normal gaussiana
# loc = media
# scale = desviacion estandar
# size = tamaño
np.random.normal(loc=100, scale=10, size=10)
```

```python
np.random.normal(loc=100, scale=10, size=1000).mean()
```

### Tres distribuciones importantes

## Normal, Uniforme y Exponencial

```python
datos = np.random.normal(size=1000000)
plt.hist(datos, color="pink",ec="red")
plt.title("Distribución Normal")
plt.show()
```

```python
datos = np.random.uniform(size=1000000)
plt.hist(datos, color="skyblue",ec="black")
plt.title("Distribución Uniforme")
plt.show()
```

```python
datos = np.random.exponential(size=1000000)
plt.hist(datos,color="green",ec="black")
plt.title("Distribución Exponencial")
plt.show()
```
